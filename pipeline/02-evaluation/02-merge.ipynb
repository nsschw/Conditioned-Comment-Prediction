```python
import pandas as pd
```

```python
df = pd.read_csv("../../outputs/results.csv")
```

```python
df["qwen3_embedding_distance"] = df["qwen3_embedding_distance"].apply(lambda x: eval(x)["mean_distance"])
df["gemma_embedding_distance"] = df["gemma_embedding_distance"].apply(lambda x: eval(x)["mean_distance"])
df["luxembourgish_embedding_distance"] = df["luxembourgish_embedding_distance"].apply(lambda x: eval(x)["mean_distance"])
```

```python
# drop as4 runs
length_before = len(df)
df = df[df["data_path"].str.contains("as4") == False].reset_index(drop=True)
length_after = len(df)
print(f"Dropped {length_before - length_after} as4 runs")
```

```python
df.head()
```

```python
df = df.groupby(["model_name", "data_path"]).agg(
    bleu_mean=("bleu", "mean"),
    bleu_std=("bleu", "std"),
    bleu_precision1_mean=("bleu_precision_1", "mean"),
    bleu_precision1_std=("bleu_precision_1", "std"),
    bleu_precision2_mean=("bleu_precision_2", "mean"),
    bleu_precision2_std=("bleu_precision_2", "std"),
    bleu_precision3_mean=("bleu_precision_3", "mean"),
    bleu_precision3_std=("bleu_precision_3", "std"),
    bleu_precision4_mean=("bleu_precision_4", "mean"),
    bleu_precision4_std=("bleu_precision_4", "std"),
    length_ratio_mean=("length_ratio", "mean"),
    length_ratio_std=("length_ratio", "std"),

    rouge1_mean=("rouge1", "mean"),
    rouge1_std=("rouge1", "std"),
    rouge2_mean=("rouge2", "mean"),
    rouge2_std=("rouge2", "std"),
    rougel_mean=("rougeL", "mean"),
    rougel_std=("rougeL", "std"),

    qwen3_embedding_distance_mean=("qwen3_embedding_distance", "mean"),
    qwen3_embedding_distance_std=("qwen3_embedding_distance", "std"),

    gemma_embedding_distance_mean=("gemma_embedding_distance", "mean"),
    gemma_embedding_distance_std=("gemma_embedding_distance", "std"),

    luxembedder_embedding_distance_mean=("luxembourgish_embedding_distance", "mean"),
    luxembedder_embedding_distance_std=("luxembourgish_embedding_distance", "std"),

).reset_index()
```
