```python
import pandas as pd
import json
```

```python
llama = "../../outputs/Llama-3.1-8B-Instruct/predictions_20251201_222421.json"
llama_ft = "../../outputs/Llama-3.1-8B-Instruct-ger-bio-history/predictions_20251128_131538.json"
```

```python
with open(llama) as f:
    llama_data = json.load(f)

with open(llama_ft) as f:
    llama_ft_data = json.load(f)
```

```python
prompts = [elem["prompt"] for elem in llama_data["generations"]]
base_model_reply = [elem["prediction"] for elem in llama_data["generations"]]
authentic_reply = [elem["reference"] for elem in llama_data["generations"]]
ft_model_reply = [elem["prediction"] for elem in llama_ft_data["generations"]]
df = pd.DataFrame({
    "prompt": prompts,
    "base_model_reply": base_model_reply,
    "ft_model_reply": ft_model_reply,
    "authentic_reply": authentic_reply
})
df.to_csv("llama3.1-8b_german-comments.csv", index=False)
```
