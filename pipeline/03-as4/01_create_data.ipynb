```python
from echo.eval.evaluate import evaluate
from pathlib import Path
import pandas as pd
import gc
import torch
import tqdm
import json
```

```python
testing_data = json.load(open("../../data/processed/eng_test_history.json"))
```

```python
def filter_data(data, length = 3):
    filtered_data = []
    for dialog in data:
        if ((len(dialog["messages"]) - 2) / 2) >= length:
            new_messages = [dialog["messages"][0]]
            new_messages.extend(dialog["messages"][-1*((length*2)+1):])
            filtered_data.append({"messages": new_messages, "human_comment": dialog["human_comment"]})
    return filtered_data
```

```python
for i in range(30):
    temp_test_data = filter_data(testing_data, length=i)
    json.dump(temp_test_data, open(f"../../data/processed/eng_test_history_as4_{i}.json", "w"))   
```
