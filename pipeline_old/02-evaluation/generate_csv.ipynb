```python
import pandas as pd
import json
```

```python
qwen = "../../outputs/Qwen3-8B/predictions_20251124_160948.json"
qwen_ft = "../../outputs/Qwen3-8B-lux-history/predictions_20251123_160203.json"

mistral = "../../outputs/Ministral-8B-Instruct-2410/predictions_20251124_111432.json"
mistral_ft = "../../outputs/Ministral-8B-Instruct-2410-lux-history/predictions_20251123_144640.json"

llama = "../../outputs/Llama-3.1-8B-Instruct/predictions_20251124_095708.json"
llama_ft = "../../outputs/Llama-3.1-8B-Instruct-lux-history/predictions_20251123_133351.json"
```

```python
with open(qwen) as f:
    qwen_data = json.load(f)

with open(qwen_ft) as f:
    qwen_ft_data = json.load(f)

with open(mistral) as f:
    mistral_data = json.load(f)

with open(mistral_ft) as f:
    mistral_ft_data = json.load(f)

with open(llama) as f:
    llama_data = json.load(f)

with open(llama_ft) as f:
    llama_ft_data = json.load(f)
```

```python
# check that data is the same for qwen

str(mistral_data["generations"][0]["prompt"]) == str(mistral_ft_data["generations"][0]["prompt"])
```

```python
prompts = [elem["prompt"] for elem in qwen_data["generations"]]
base_model_reply = [elem["prediction"] for elem in qwen_data["generations"]]
authentic_reply = [elem["reference"] for elem in qwen_data["generations"]]
ft_model_reply = [elem["prediction"] for elem in qwen_ft_data["generations"]]

df = pd.DataFrame({
    "prompt": prompts,
    "base_model_reply": base_model_reply,
    "ft_model_reply": ft_model_reply,
    "authentic_reply": authentic_reply
})
df.to_csv("qwen_rtl-comments.csv", index=False)
```
